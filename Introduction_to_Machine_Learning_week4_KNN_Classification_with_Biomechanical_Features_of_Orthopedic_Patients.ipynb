```python
x,y = data.loc[:,data.columns != 'class'], data.loc[:,'class']
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 3) #KNN kullanyoruz
knn.fit(x,y)
prediction = knn.predict(x)
print('Prediction: {}'.format(prediction))
```

```python
# train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state = 1)
```

```python
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(x_train,y_train)
prediction = knn.predict(x_test)
#print('Prediction: {}'.format(prediction))
print('With KNN (K=3) accuracy is: ',knn.score(x_test,y_test)) # accuracy
#knn deerini hesaplayp elde ettik
```

```python
test_accuracy = []
train_accuracy = []
n = np.arange(1,50)
for i,k in enumerate(n):
  knn1 = KNeighborsClassifier(n_neighbors = k)
  knn1.fit(x_train,y_train)
  train_accuracy.append(knn1.score(x_train,y_train))
  test_accuracy.append(knn1.score(x_test,y_test))
```

```python
# Plot
plt.figure(figsize=[13,8])
plt.plot(n, test_accuracy, label = 'Testing Accuracy')
plt.plot(n, train_accuracy, label = 'Training Accuracy')
plt.legend()
plt.title('-value VS Accuracy')
plt.xlabel('Number of Neighbors')
plt.ylabel('Accuracy')
plt.xticks(n)
plt.savefig('graph.png')
plt.show()
print("Best accuracy is {} with K = {}".format(np.max(test_accuracy),1+test_accuracy.index(np.max(test_accuracy))))
best_n_value = 1+test_accuracy.index(np.max(test_accuracy))
```

```python
knn = KNeighborsClassifier(n_neighbors = best_n_value)
knn.fit(x_train,y_train)
print('With KNN k={} accuracy is: {}'.format(best_n_value,knn.score(x_test,y_test))) # accuracy
```

```python
y_pred = knn.predict(x_test)
y_true = y_test
#%% confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_true,y_pred)


# %% cm visualization
import seaborn as sns
import matplotlib.pyplot as plt

f, ax = plt.subplots(figsize =(5,5))
sns.heatmap(cm,annot = True,linewidths=0.5,linecolor="w",fmt = ".0f",ax=ax, cbar=False)
plt.xlabel("y_pred")
plt.ylabel("y_true")
plt.show()
```
